```python
import pandas as pd
```

```python
beats = pd.read_csv("cpdb_public_data_offrassignmentattendance.csv")
```

```python
beats.dropna(inplace=True)
beats.officer_id = beats.officer_id.astype(int)
beats.beat_id = beats.beat_id.astype(int)
```

```python
beat_nums = set()
import json
with open("bounds.geojson") as f:
    gj = json.load(f)
for feature in gj['features']:
    beat_nums.add(int(feature['properties']['beat_num']))
```

```python
beat_nums
```

```python
beats = beats[beats['beat_id'].isin(beat_nums)] 
```

```python
beats_mode = beats.groupby('officer_id')['beat_id'].apply(pd.Series.mode).to_frame()
```

```python
beats_mode.nunique()
```

```python
officer_data = pd.read_csv("cpdb_public_data_officer.csv")
```

```python
officer_data
```

```python
officer_data = officer_data.dropna()
```

```python
df = beats_mode.merge(officer_data, left_on='officer_id', right_on='id', how='left')
```

```python
df = df.dropna()
```

```python
df
```

```python
df.groupby("beat_id")['civilian_allegation_percentile','honorable_mention_percentile','internal_allegation_percentile','trr_percentile','allegation_count','sustained_count','civilian_compliment_count'].mean()
```
